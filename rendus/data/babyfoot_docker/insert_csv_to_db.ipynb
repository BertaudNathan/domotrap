```python
import pandas as pd
import sqlite3
import os

conn = sqlite3.connect("data/BABYFOOT.db")

path = os.path.join(os.getcwd(), "data", "babyfoot_clean.csv")
df = pd.read_csv(path, low_memory=False)
print(f"📦 CSV chargé : {len(df)} lignes et {len(df.columns)} colonnes")
print("Colonnes disponibles :", df.columns.tolist(), "\n")

df_player = df[['player_id', 'player_name', 'player_canonical_name']].copy()
df_player.rename(columns={'player_canonical_name': 'player_pseudo'}, inplace=True)
df_player['player_password'] = 'default_password'
df_player['player_birthday'] = None

df_player.drop_duplicates(subset=['player_id'], inplace=True)
print(f"👤 {len(df_player)} joueurs uniques prêts à insérer")

if not df_player.empty:
    df_player.to_sql("player", conn, if_exists="append", index=False)
    print("✅ Insertion réussie dans player")
else:
    print("⚠️ Aucun joueur à insérer — vérifie player_id")

count = conn.execute("SELECT COUNT(*) FROM player").fetchone()[0]
print(f"📊 Total actuel dans player : {count} lignes")

conn.close()
print("💾 Connexion fermée")
```

```python
import pandas as pd
import sqlite3
import os

# ===============================================
# 1️⃣ Connexion à la base SQLite
# ===============================================
db_path = os.path.join(os.getcwd(), "data", "BABYFOOT.db")
conn = sqlite3.connect(db_path)

# ===============================================
# 2️⃣ Chargement du CSV nettoyé
# ===============================================
csv_path = os.path.join(os.getcwd(), "data", "babyfoot_clean.csv")
df = pd.read_csv(csv_path, low_memory=False)
print(f"📦 CSV chargé : {len(df)} lignes et {len(df.columns)} colonnes")

# ===============================================
# 3️⃣ Préparation du DataFrame pour 'match_'
# ===============================================
df_match = pd.DataFrame({
    "Id_match": df["game_id"].astype(str),
    "match_date": df["game_date"].astype(str),
    "match_duration": pd.to_numeric(df["game_duration"], errors="coerce"),
    "match_final_score_red": pd.to_numeric(df["final_score_red"], errors="coerce"),
    "match_final_score_blue": pd.to_numeric(df["final_score_blue"], errors="coerce"),
    "match_winner": df["winner"].astype(str),
    "match_rating": pd.to_numeric(df["rating_raw"], errors="coerce"),
    "Id_babyfoot_table": df["table_id"].astype(str)
})

# Supprime les doublons de match_id si présents
df_match = df_match.drop_duplicates(subset=["Id_match"])
print(f"🏆 {len(df_match)} matchs prêts à être insérés")

# ===============================================
# 4️⃣ Insertion dans SQLite
# ===============================================
try:
    df_match.to_sql("match_", conn, if_exists="append", index=False)
    print("✅ Insertion réussie dans match_")
except Exception as e:
    print(f"❌ Erreur lors de l'insertion : {e}")

# ===============================================
# 5️⃣ Vérification du total
# ===============================================
count = conn.execute("SELECT COUNT(*) FROM match_").fetchone()[0]
print(f"📊 Total actuel dans match_ : {count} lignes")

conn.close()
print("💾 Connexion fermée")

```

```python
import pandas as pd
import sqlite3
import os

# ===============================================
# 1️⃣ Connexion à la base SQLite
# ===============================================
db_path = os.path.join(os.getcwd(), "data", "BABYFOOT.db")
conn = sqlite3.connect(db_path)

# ===============================================
# 2️⃣ Chargement du CSV nettoyé
# ===============================================
csv_path = os.path.join(os.getcwd(), "data", "babyfoot_clean.csv")
df = pd.read_csv(csv_path, low_memory=False)
print(f"📦 CSV chargé : {len(df)} lignes et {len(df.columns)} colonnes")

# ===============================================
# 3️⃣ Préparation du DataFrame pour 'play'
# ===============================================
df_play = pd.DataFrame({
    "Id_match": df["game_id"].astype(str),               # même format que match_.Id_match
    "player_id": df["player_id"].astype(str),            # même format que player.player_id
    "role": df["player_role"].astype(str),
    "goal": pd.to_numeric(df["player_goals"], errors="coerce"),
    "own_goal": pd.to_numeric(df["player_own_goals"], errors="coerce"),
    "mood": pd.to_numeric(df["mood"], errors="coerce"),
    "comment": df["player_comment"].astype(str),
    "team_color": df["team_color"].astype(str),
    "is_substitute": df["is_substitute"].apply(lambda x: bool(x) if pd.notna(x) else None)
})

# Suppression des lignes sans Id_match ni player_id
df_play = df_play.dropna(subset=["Id_match", "player_id"])
df_play = df_play.drop_duplicates(subset=["Id_match", "player_id"])  # éviter les doublons
print(f"🎮 {len(df_play)} lignes prêtes à être insérées dans play")

# ===============================================
# 4️⃣ Insertion dans SQLite
# ===============================================
try:
    df_play.to_sql("play", conn, if_exists="append", index=False)
    print("✅ Insertion réussie dans play")
except Exception as e:
    print(f"❌ Erreur lors de l'insertion : {e}")

# ===============================================
# 5️⃣ Vérification du total
# ===============================================
count = conn.execute("SELECT COUNT(*) FROM play").fetchone()[0]
print(f"📊 Total actuel dans play : {count} lignes")

conn.close()
print("💾 Connexion fermée")
```
